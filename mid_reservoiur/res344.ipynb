```python
import os
import matplotlib.pyplot as plt
#import seaborn as sns

from shapely.geometry import Point
import pandas as pd
import geopandas as gpd
from geopandas import GeoSeries,GeoDataFrame
import requests
import geojson
import numpy as np
```

```python
# อ่านค่า CSV อ่างเก็บน้ำขนาดกลาง 344 แห่ง
data_mid_rsv_now=pd.read_csv(r'./344Data/M25092019.csv',encoding='utf-8')
data_mid_rsv_now
```

```python
#อ่านข้อมูล shape file  สำหรับตั้งต้นเพื่อให้ นำตารางจากก่อนหน้านี้มา join
m344=gpd.read_file(r'./350shp/350M_rg_v2.shp',encoding='UTF-8')
m344
```

```python
# merge ข้อมูล จะได้ shape ที่มีข้อมูลสถานะอ่างกลาง
data_m344=m344.merge(data_mid_rsv_now,on='nwcc_id')
#data_m344=data_mid_rsv_now.merge(m344,on='nwcc_id')
data_m344
```

```python
# เติมข้อมูล
# วิเคราะห์ Gap Remain สามารถรับน้ำได้อีก
data_m344['Gap_Remain']=data_m344['rsv_cap']-data_m344['rsv_vol']

#วิเคราะห์ กรณีน้อยกว่า 15
data_m344['lt_15']= np.where(data_m344['rsv_vol_percent']<15, 1,0)
#วิเคราะห์ กรณีน้อยกว่า 15 และใส่ชื่ออ่าง ฯ ร่วม
data_m344['lt_15_name']=data_m344['lt_15'].astype(str) + '-' + data_m344['rsv_name']

#วิเคราะห์ กรณีน้อยกว่า 30
data_m344['lt_30']= np.where(data_m344['rsv_vol_percent']<30, 1,0)
#วิเคราะห์ กรณีน้อยกว่า 30 และใส่ชื่ออ่าง ฯ ร่วม
data_m344['lt_30_name']=data_m344['lt_30'].astype(str) + '-' + data_m344['rsv_name']

#วิเคราะห์ราย กรณีมากกว่า 30 น้อยกว่า 60
data_m344['bt_30_60']= np.where((data_m344['rsv_vol_percent']>=30) & (data_m344['rsv_vol_percent']<60), 1,0)

#วิเคราะห์ กรณีมากกว่า 60 น้อยกว่า 80
data_m344['bt_60_80']= np.where((data_m344['rsv_vol_percent']>=60) & (data_m344['rsv_vol_percent']<80), 1,0)

#วิเคราะห์ กรณีมากกว่า 80
data_m344['gt_80']= np.where((data_m344['rsv_vol_percent']>=80), 1,0)
#วิเคราะห์ กรณีมากกว่า 80 และใส่ชื่ออ่าง ฯ ร่วม
data_m344['gt_80_name']=data_m344['gt_80'].astype(str) + '-' + data_m344['rsv_name']

data_m344
```

```python
# save data ลง csv ใน result/result344M_[วันที่].csv
data_m344[['nwcc_id','name','tambol','amphoe','province','Region_4','rg_onwr62','rg_moi62','rg_sk01','Basin_N','rsv_cap','rsv_vol','rsv_inflow','rsv_outflow','lt_15','lt_30','bt_30_60','bt_60_80','gt_80','Gap_Remain']].to_csv(r'./result/result350M_26092019.csv',encoding='utf-8')
#data_m344[['nwcc_id','name','tambol','amphoe','province','Region_4','Basin_N','rsv_cap','rsv_vol','rsv_inflow','rsv_outflow','lt_15','lt_30','bt_30_60','bt_60_80','gt_80','Gap_Remain']].to_csv(r'./result/result344M_12072019.csv',encoding='utf-8')
# save data เป็น shape file แบบ point ที่ result/result344M_[วันที่].shp

gdf= gpd.GeoDataFrame(data_m344,geometry=[Point(x, y) for x, y in zip(data_m344.lng, data_m344.lat)])
gdf.to_file("./result/result350M_26092019.shp",encoding='utf-8')
```

```python
# สร้างฟังก์ชั่น เพื่อเก็บชื่อ
def returnname(x):
    s=[]
    for xx in x:
        arrayxx= xx.split('-')
        if arrayxx[0]=="1":
            s.append(arrayxx[1])
    return ",".join(s)
a=["0-asdf","0-bbbb","1-cccc","0-dddd","1-eeee"]
returnname(a)
```

```python
# ตัดข้อมูลที่ใช้เท่าที่จำเป็น
m344_agg=data_m344[['rg_onwr62','rg_moi62','rg_sk01','Basin_N','province','name','rsv_vol','rsv_vol_percent','nwcc_id','Gap_Remain','lt_15','lt_15_name','lt_30','lt_30_name','bt_30_60','bt_60_80','gt_80','gt_80_name']]
#m344_agg=data_m344[['Region_4','Basin_N','province','name','rsv_vol','rsv_vol_percent','nwcc_id','Gap_Remain','lt_15','lt_15_name','lt_30','lt_30_name','bt_30_60','bt_60_80','gt_80','gt_80_name']]
m344_agg
```

```python
#ทำ ข้อมูลสรุป
d = {'nwcc_id': ['count'],'rsv_vol': ['sum'],'Gap_Remain': ['sum'] ,'lt_15': ['sum'],'lt_15_name':lambda x: returnname(x), 'lt_30': ['sum'], 'bt_30_60': ['sum'], 'bt_60_80': ['sum'], 'gt_80': ['sum'],'gt_80_name':lambda x: returnname(x)}
result_basin=m344_agg.groupby(['rg_onwr62','Basin_N',],as_index=True).agg(d)
#result_basin=m344_agg.groupby(['Region_4','Basin_N',],as_index=True).agg(d)
result_basin
```

```python
result_province=m344_agg.groupby(['rg_sk01','province',],as_index=True).agg(d)
result_province
```

```python
#ทำ ข้อมูลสรุป น้อยกว่า 15 และ น้อยกว่า 30
agg_def = {'nwcc_id': ['count'],'rsv_vol': ['sum'],'Gap_Remain': ['sum'] ,'lt_15': ['sum'],'lt_15_name':lambda x: returnname(x), 'lt_30': ['sum'],'lt_30_name':lambda x: returnname(x)}
result_lt_15_30=m344_agg.groupby(['rg_sk01','province',],as_index=True).agg(agg_def)
result_lt_15_30
```

```python
# save all to excel
#data=data_m344[['nwcc_id','name','tambol','amphoe','province','rg_onwr62','rg_moi62','Basin_N','rsv_cap','rsv_vol','rsv_inflow','rsv_outflow','lt_15','lt_30','bt_30_60','bt_60_80','gt_80','Gap_Remain']]
data=data_m344[['nwcc_id','name','tambol','amphoe','province','rg_moi62','rg_onwr62','rg_sk01','Basin_N','rsv_cap','rsv_vol','rsv_inflow','rsv_outflow','lt_15','lt_30','bt_30_60','bt_60_80','gt_80','Gap_Remain']]
with pd.ExcelWriter('./result/result_m350.xlsx') as writer:
    result_basin.to_excel(writer,sheet_name="result_basin")
    result_province.to_excel(writer,sheet_name="result_province")
    result_lt_15_30.to_excel(writer,sheet_name="result_lt_15_30")
    data.to_excel(writer,sheet_name="data")
```
